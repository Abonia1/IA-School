```python
%matplotlib inline
import pandas as pd
import folium
from folium.plugins import MarkerCluster
```

```python
listings = pd.read_csv('C:/Users/ABONIA/Downloads/listings.csv.gz', compression='gzip')
```

```python
mapData=listings[['latitude','longitude','id','street','zipcode','city','state','smart_location']]
```

```python
locations = mapData[['latitude', 'longitude']]
locationlist = locations.values.tolist()
len(locationlist)
#locationlist[7]
```

```python
location_10=locationlist[:10]
location_200=locationlist[:200]
location_500=locationlist[:500]
```

```python
map = folium.Map(location=[48.8566,2.3522], zoom_start=12)
for point in range(0, len(location_10)):
    folium.Marker(location_10[point], popup=mapData['id'][point]).add_to(map)
map
```

```python
#openstreet map to point listing id based on location



map2 = folium.Map(location=[48.8566,2.3522], tiles='CartoDB positron', zoom_start=11)

marker_cluster = MarkerCluster().add_to(map2)

for point in range(0, len(location_200)):
    folium.Marker(location_200[point], popup=mapData['id'][point], icon=folium.Icon(icon_color='white', icon='home', angle=0, prefix='fa')).add_to(map2)
map2
```

```python
map2 = folium.Map(location=[48.8566,2.3522], tiles='Stamen Terrain', zoom_start=12)

marker_cluster = MarkerCluster().add_to(map2)

for point in range(0, len(location_200)):
    folium.Marker(location_200[point], popup=mapData['id'][point], icon=folium.Icon(color='darkblue', icon_color='white', icon='male', angle=0, prefix='fa')).add_to(marker_cluster)
map2
```
