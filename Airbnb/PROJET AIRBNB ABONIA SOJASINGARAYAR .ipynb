```python
#Getting number of listing by neighbourhood
neighbour_listings = listings[['id','neighbourhood']].groupby('neighbourhood').count().round().sort_values(by='id', ascending=False)
neighbour_listings.reset_index(0, inplace=True)
print(neighbour_listings.head())


plt.figure(figsize=(15,12))
sns.barplot(x=neighbour_listings['id'], y=neighbour_listings['neighbourhood'], palette="rocket")
plt.xlabel('Number of listings_id ', fontsize=15)
plt.ylabel('Neighbourhood ', fontsize=15)
plt.title("Number of listings by neighbourhood", fontsize=20)
plt.xticks(rotation= 45)
plt.tight_layout()
```

```python
propertyType_listings = listings[['id','property_type']].groupby('property_type').count().round().sort_values(by='id', ascending=False)
propertyType_listings.reset_index(0, inplace=True)
propertyType_listings.head()

roomType_listings = listings[['id','room_type']].groupby('room_type').count().round().sort_values(by='id', ascending=False)
roomType_listings.reset_index(0, inplace=True)
roomType_listings.head()

pd.concat([propertyType_listings,roomType_listings],axis=1)
```

```python
propertyType_listings = listings[['id','property_type']].groupby('property_type').mean().round().sort_values(by='id', ascending=False)
propertyType_listings.reset_index(0, inplace=True)
print(propertyType_listings.head())


plt.figure(figsize=(10,8))
sns.barplot(x= propertyType_listings['id'], y=propertyType_listings['property_type'], palette="rocket")
plt.xlabel('Number of listings', fontsize=15)
plt.ylabel('Property Type ', fontsize=15)
plt.title("Property type in Paris", fontsize=20)
plt.xticks(rotation= 90)
plt.tight_layout()
```

```python
#Getting  price and accomodation
person_accomodationdf = pd.DataFrame({'accommodates':listings['accommodates'],'price':listings['price'],'id':listings['id']})
person_accomodationdf.head()
```

```python
#2 person accomodation
accommodates = [2]
acc=person_accomodationdf[person_accomodationdf.accommodates.isin(accommodates)]
acc.head()
```

```python
#Convert price to float type
acc['price']=acc['price'].str.replace(',', '').str.replace('$', '').astype(float)
acc.head()
```

```python
#Get price between 0 and 140 to plot
price=acc['price'].between(0, 140.00, inclusive=False)
price.shape
```

```python
#Insert the price selected between 0 and 140
price_140=acc[price]
price_140
```

```python
#Convert dollar to euro
price_140['price']=price_140['price']/1.08
price_140
```

```python
#Get average price
price_person = price_140[['accommodates','price','id']].groupby('price').mean().round().sort_values(by='id', ascending=False)
price_person.reset_index(0, inplace=True)
price_person.shape
```

```python
#Plot 2 person accomodation vs price
plt.figure(figsize=(18,20))
sns.barplot(x=price_person['price'], y=price_person['id'],orient="h")
plt.xlabel('Average daily price-Euro', fontsize=15)
plt.ylabel('ID', fontsize=15)
plt.title("Average daily price for 2-persons accomodation", fontsize=20)
plt.xticks(rotation= 90)
plt.tight_layout()
```

```python
reviews=pd.read_csv("C:/Users/ABONIA/Downloads/reviews.csv")
reviews.head()
```

```python
calendar = pd.read_csv('C:/Users/ABONIA/Downloads/calendar.csv.gz', compression='gzip')
calendar.head()
```

```python
#time vs listing
start_date = '2019-01'
end_date = '2019-12'
mask = (reviews['date'] > start_date) & (reviews['date'] <= end_date)


df = reviews.loc[mask]
df['Date'] = pd.to_datetime(df["date"])

df=df[df['Date'].dt.month.isin([1,5, 7,9,10])]

#type_df = review.groupby('listing_id').id.count()
type_df =df[['listing_id','Date']].groupby('Date').count()
type_df.reset_index(0, inplace=True)
#type_df = type_df.reset_index()
#type_df = type_df.rename(columns={'date': 'number_of_listings'})
type_df
```

```python
from matplotlib.dates import MonthLocator, DateFormatter
plt.plot(type_df['Date'],type_df['listing_id'])

ax = plt.gca()


mnt = MonthLocator(bymonthday=1, interval=2)
mnt_fmt = DateFormatter("%b '%y")
ax.xaxis.set_major_locator(mnt)
ax.xaxis.set_major_formatter(mnt_fmt)

ax.grid( linestyle='-', linewidth=1)
#ax.grid(color='r', linestyle='-', linewidth=2)

plt.ylabel('Number of listings available', fontsize=18)
plt.xlabel('Date', fontsize=18)
ax.set_title('Number of listings available by date', fontsize=20)
ax.figure.set_size_inches(20, 10)

plt.show()
```
