```python
room_type_df = listings.groupby('room_type').id.count()
room_type_df = room_type_df.reset_index()
room_type_df = room_type_df.rename(columns={'id': 'number_of_listings'})
room_type_df
```

```python
sns.barplot(x=arr['id'], y=arr['neighbourhood'], palette="rocket")
plt.xlabel('listings_id ', fontsize=15, color='Green')
plt.ylabel('neighbourhood ', fontsize=15, color='Green')
plt.title("listings by neighbourhood", fontsize=18, color='Red')
plt.xticks(rotation= 45)
plt.tight_layout()
```

```python
from collections import Counter
room = listings.room_type
r = Counter(room)

# get the dataframe with room count
room_df = pd.DataFrame.from_dict(r, orient='index').sort_values(by=0)
room_df.columns = ['room_type']

# Plot using distribution diagram to find the distribution of rooms 
room_df.plot.pie(y = 'room_type', 
                 colormap = 'Blues_r',
                 figsize = (12, 12),
                 fontsize = 16, 
                 autopct = '%.2f',
                 legend = False,
                 title = 'Distribution Diagram')
```
