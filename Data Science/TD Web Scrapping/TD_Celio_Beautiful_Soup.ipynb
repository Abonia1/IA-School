```python
### listes pour chaque colonne du dataframe
magasin_liste = []
ville_liste = []
adresse_liste = []
code_postal_liste = []

###fonction pour récupérer les adresses de chaque magasin français à partir de la liste finale href_raw_5
def nom_magasin():
    for i in href_raw5:
        src = requests.get(i).text
        soup = BeautifulSoup(src, 'lxml')
        liste = soup.find('div', class_='Information')
        nom_magasin = liste.h3.text
        nom_magasin_clean = re.sub(r'\n', ' ', nom_magasin).strip()
        magasin_liste.append(nom_magasin_clean)
    return None


def nom_ville():
    for i in href_raw5:
        src = requests.get(i).text
        soup = BeautifulSoup(src, 'lxml')
        liste = soup.find('div', class_='Information')
        ville = liste.find('span', class_='city').text
        ville_liste.append(ville)
    return None

def code_postal():
    for i in href_raw5:
        src = requests.get(i).text
        soup = BeautifulSoup(src, 'lxml')
        liste = soup.find('div', class_='Information')
        cp = liste.find('span', itemprop='postalCode').text
        code_postal_liste.append(cp)
    return None    

def adresse():
    liste_adresse_temp = []
    for i in href_raw5:
        src = requests.get(i).text
        soup = BeautifulSoup(src, 'lxml')
        liste = soup.find('div', class_='Information')
        adresse = liste.find('span', itemprop='streetAddress').text
        adresse_liste.append(re.sub(r'\r\n', '', adresse).strip())
    return None
```

```python
### création d'un datframe à partir de toutes les  listes
celio_df = pd.DataFrame({
    'Nom_magasin': [m for m in magasin_liste],
    'Ville' : [v for v in ville_liste],
    'Adresse': [a for a in adresse_liste],
    'Code_postal': [c for c in code_postal_liste]
})
```
